## Creating or transforming variables → Feature Engineering
## Generating new columns → Feature Creation

### "Ship Mode" column has values like "Not Available" and "unknown" so while reading data from file, we can handle them as NA values

```python
#hansling null values 
# by default it considers few values as NA, but I want to it consider Not-Available, unkowon as NA only
import numpy as np

df1 = df.copy()
df1['Ship Mode'] = df1['Ship Mode'].replace(['Not Available', 'unknown'], np.nan)

# Check unique values
print(df1['Ship Mode'].unique())
```

```python
df1.head()
```

# We see the "Order Date" columns  datatype is object, So we need to change it into datetime

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], dayfirst=True)
```

```python
df1.info()
```

### Derive new columns "Selling Price" from list price and discount price [Feature Engineering]

```python
df.columns
```

```python
df1['Selling Price'] = df1['List Price'] - df1['List Price'] * (df1['Discount Percent'] /100)
df1['Selling Price']
```

### Derive new columns "Profit" from list "Selling Price" and "cost price" [Feature Engineering]

```python
df1['Profit'] = df1['Selling Price'] - df1['cost price']
df1['Profit']
```

```python
df1.head(5)
```

### Now let's rename the Columns for understandable 

```python
df1.rename(columns = {'Profit' : 'Unit_Profit' , 'Selling Price' : 'Unit_Selling_Price'}, inplace=True) 
```

```python
df1.head(5)
```

### Now let's drop some columns which not necessary

```python
df1.drop(columns = ['cost price', 'List Price', 'Discount Percent'], inplace = True)
```

```python
df1.head(2)
```

# Let's connect and Export the files from Jupyter notebook to SSMS 

```python
from sqlalchemy import create_engine, text

# Server and DB info
server = r'OMOR\SQLEXPRESS'  # Use your actual SQL Server instance
database = 'PythonNSqlProjects'     # Replace with your actual database name

# Create engine
engine = create_engine(f'mssql+pyodbc://{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server')

# Test connection
with engine.connect() as conn:
    result = conn.execute(text("SELECT GETDATE()"))  # 👈 Wrap query in text()
    for row in result:
        print(row)
```

```python
df1.to_sql(
    'orders_data',
    con=engine,        # use the engine, not the closed conn
    index=False,
    if_exists='replace'  # or 'append'
)
```
